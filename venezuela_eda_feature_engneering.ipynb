## Cleaning the values

```python
df.isna().mean()*100
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
interpol_cols=['gdp_growth_pct','oil_rents_pct_gdp',
                'total_natural_resource_rents_pct_gdp',
       'fuel_exports_pct_merch_exports',
       'ores_and_metals_exports_pct_merch_exports']
df[interpol_cols]=df[interpol_cols].interpolate(method='linear')
df[interpol_cols]=df[interpol_cols].ffill().bfill()
```

```python
df=df.drop(columns=['gdp_growth_5yr_avg'])
```

```python
df['gdp_growth_5yr_avg'] = df['gdp_growth_pct'].rolling(5).mean()
```

cleaned the values using linear interpolation as it was a time-series dataset, Understood the volatility of Venezuela's economy and overdependence on natural resources. Dropped the derived column and created it later with clean values to avoid error propagation.

```python
df=df.sort_values('year')
```

# feature engineering


```python
for lag in [1,2,3]:
    df[f'gdp_growth_{lag}']=df['gdp_growth_pct'].shift(lag)
```

```python
for lag in [1,2,3]:
    df[f'fuel_exports_pct_{lag}']=df['fuel_exports_pct_merch_exports'].shift(lag)
```

created three new columns so that the model can read the past parameters on which current year's performance depends

# saving the clean dataset for model training

```python
df.to_csv('venezuela_clean_features.csv',index=False)
```
